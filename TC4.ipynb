```python
import numpy as np
import matplotlib.pyplot as plt

#data
time = np.array([1, 2, 3, 4, 5, 6, 7])
temperature = np.array([13, 15, 20, 14, 15, 13, 10])

A = np.vstack([time, np.ones(len(time))]).T
b = temperature
x = np.linalg.lstsq(A, b, rcond=None)[0]

#At*A*x = At*b
At = A.T
At_A = np.dot(At, A)
At_b = np.dot(At, b)
x = np.linalg.solve(At_A, At_b)

#plot linear fit
plt.plot(time, temperature, 'o')
plt.plot(time, x[0]*time + x[1])
plt.xlabel('Time (minutes)')
plt.ylabel('Temperature (degrees C)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([0, 1.5, 2.5, 3.5, 4.5])
y = np.array([2, 3.6, 5.4, 8.1, 12])

A = np.vstack([x, np.ones(len(x))]).T
b = np.log(y)
k = np.linalg.lstsq(A, b, rcond=None)[0]

#At*A*x = At*b
At = A.T
At_A = np.dot(At, A)
At_b = np.dot(At, b)
k = np.linalg.solve(At_A, At_b)

#plot exponential fit
plt.plot(x, y, 'o')
plt.plot(x, np.exp(k[0]*x + k[1]))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#data
year = np.array([1950, 1955, 1960, 1965, 1966, 1967, 1968])
naccidents = np.array([8200, 9800, 10300, 13100, 13500, 13600, 14500])
accidentsvehicles = np.array([1687, 1576, 1396, 1438, 1418, 1384, 1414])

#Quadratic least-squares of the number of accidents per vehicle
A = np.vstack([year**2, year, np.ones(len(year))]).T
b = accidentsvehicles
x = np.linalg.lstsq(A, b, rcond=None)[0]

#At*A*x = At*b
At = A.T
At_A = np.dot(At, A)
At_b = np.dot(At, b)
x = np.linalg.solve(At_A, At_b)

#predict the number of accidents per vehicle in 1990
year_1990 = 1990
accidentsvehicles_1990 = x[0]*year_1990**2 + x[1]*year_1990 + x[2]
print('The number of accidents per vehicle in 1990 is', round(accidentsvehicles_1990))

year = np.append(year, year_1990)
accidentsvehicles = np.append(accidentsvehicles, accidentsvehicles_1990)

# generate a sequence of evenly spaced numbers between the minimum and maximum year values
year_seq = np.linspace(year.min(), year.max(), 100)

#plot quadratic fit
plt.plot(year, accidentsvehicles, 'o')
plt.plot(year_seq, x[0]*year_seq**2 + x[1]*year_seq + x[2])
plt.title('Number of accidents per vehicle (quadratic fit)/prediction for 1990')
plt.xlabel('Year')
plt.ylabel('Number of accidents per vehicle')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#data
t = np.array([0, 1, 2, 3, 4])
p = np.array([1, 1.8, 3.3, 6, 11])

#Expontential least-squares of the population
A = np.vstack([t, np.ones(len(t))]).T
b = np.log(p)
x = np.linalg.lstsq(A, b, rcond=None)[0]

#At*A*x = At*b
At = A.T
At_A = np.dot(At, A)
At_b = np.dot(At, b)
x = np.linalg.solve(At_A, At_b)

#predict the population in 5 years
t_5 = 5
p_5 = np.exp(x[0]*t_5 + x[1])

print('The population in 5 years is', round(p_5))

t = np.append(t, t_5)
p = np.append(p, p_5)

# generate a sequence of evenly spaced numbers between the minimum and maximum year values
t_seq = np.linspace(t.min(), t.max(), 100)

#plot exponential fit
plt.plot(t, p, 'o')
plt.plot(t_seq, np.exp(x[0]*t_seq + x[1]))
plt.title('Population (exponential fit)/prediction for 5 years')
plt.xlabel('Time (years)')
plt.ylabel('Population (millions)')
plt.show()
```
